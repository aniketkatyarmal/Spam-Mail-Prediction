```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
mail_data=pd.read_csv('../input/maildataset/mail_data.csv')
print(mail_data.shape)
print(mail_data)
```

```python
mail_data.info()
```

```python
mail_data.isnull().sum()
```

```python
mail_data.head()
```

```python
mail_data.shape
```

```python
mail_data.loc[mail_data['Category']=='spam','Category']=0
mail_data.loc[mail_data['Category']=='ham','Category']=1
```

```python
mail_data.head()
```

```python
x=mail_data['Message']
y=mail_data['Category']
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

Feature Extraction
transform text data to feature vectors for logistic regression

```python
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase='True')
x_train_features=feature_extraction.fit_transform(x_train)
#dont fit x_test because we dont want to give our test data to model
x_test_features=feature_extraction.transform(x_test)

y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
print(x_train_features)
```

```python
model= LogisticRegression()
```

```python
model.fit(x_train_features,y_train)
```

**Predict on training data to confirm weather Overfitting occurs or not.**

```python
#prediction on training data

prediction_on_training_data=model.predict(x_train_features)
accuracy_on_training_data= accuracy_score(y_train,prediction_on_training_data)
print('Accuracy on training data: ',accuracy_on_training_data)
```

```python
#prediction on test data

prediction_on_test_data=model.predict(x_test_features)
accuracy_on_test_data= accuracy_score(y_test,prediction_on_test_data)
print('Accuracy on test data: ',accuracy_on_test_data)
```

```python
input_mail=[''' free fre'''] 
input_mail_features=feature_extraction.transform(input_mail)

prediction_input=model.predict(input_mail_features)
print(prediction_input)


if prediction_input[0]==1:
    print("Human mail")
else:
    print("Spam mail")
```
